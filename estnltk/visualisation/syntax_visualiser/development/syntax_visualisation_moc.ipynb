````python
def css():
        with open("syntax_visualiser_css.css") as css_file:
            contents = css_file.read()
            output = ''.join(["<style>\n", contents, "</style>"])
        return output
````

````python
def event_handler_code():
        with open("syntax_visualiser_js.js") as js_file:
            contents = js_file.read()
            output = ''.join(["<script>\n", contents, "</script>"])
            #output = ''.join(["<script>\n var text_id=", str(self._text_id),"\n", contents, "</script>"])
        return output
````

````python
def table_column(layers, attribute, deprel, index = None):
    cellid = 0
    sentence = 0
    word = 0
    if index is not None:
        table_elements = ["<td><table><tr><th>", attribute, str(index), "</th></tr>"]
    else:
        table_elements = ["<td><table><tr><th>", attribute, "</th></tr>"]
    if attribute == "feats":
        for element in getattr(layers, attribute):
            table_elements.extend(["<tr><td id = \"", str(attribute), str(index), ";", str(cellid), "\">"])
            cellid +=1
            if element is not None:
                for key in element:
                    table_elements.extend([str(key), " "])
            else:
                table_elements.append(str(element))
            table_elements.append("</td></tr>")
    elif attribute == "deprel":
        for element in getattr(layers, attribute):
            table_elements.extend(["<tr><td id = \"", str(attribute), str(index), ";", str(cellid), "\">"])
            cellid += 1
            table_elements.extend(["<select id = \"deprel", str(index), ";", str(cellid), "><option value=\"", str(cellid), ";0>", str(element), "</option>"])
            for deprel_element in deprel:
                if element != deprel_element:
                    table_elements.extend(["<option value=", str(cellid), ";", deprel_element, ">", str(deprel_element), "</option>"])
            table_elements.extend(["</select>", "</td></tr>"])
    elif attribute == "head":
        for element in getattr(layers, attribute):
            if word == len(textmalt.sentences[sentence]):
                word = 0
                sentence += 1
            table_elements.extend(["<tr><td id = \"", str(attribute), str(index), ";", str(cellid), "\">"])
            cellid += 1
            #table_elements.extend(["<select id = \"head", str(index), ""\"><option>", str(element), "</option>"])
            table_elements.extend(["<select id=\"head", str(index), ";", str(cellid)," onchange=\"get_select_value();\"><option value=\"", str(cellid), ";0\">", str(element), "</option>"])
            #table_elements.extend(["<select id=\"head", str(cellid),"\"><option value=\"0\">", str(element), "</option>"])
            for i in range(len(textmalt.sentences[sentence])):
                table_elements.extend(["<option value=", str(cellid), ";", str(i + 1), ">", str(i + 1), ": ", str(textmalt.sentences[sentence].text[i]), "</option>"])
            word += 1
            table_elements.append("</select></td></tr>")
    else:
        for element in getattr(layers, attribute):
            table_elements.extend(["<tr><td id = \"", str(attribute), str(index), ";", str(cellid), "\">", str(element), "</td></tr>"])
            cellid += 1
    table_elements.append("</table></td>")
    return table_elements
````

````python
def table(layer, attributes, deprel):
    table_elements = [css(), "<table><tr>"]
    for attribute in attributes:
        table_elements.extend(table_column(layer, attribute, deprel))
    table_elements.append("</tr></table>")
    return "".join(table_elements)
````

````python
attributes = ["text", "id", "lemma", "upostag", "xpostag", "feats", "head", "deprel", "deps", "misc"]
deprel = ["@ADVL", "@FCV", "ROOT", "@SUBJ"] #test deprel list
#display_html(table(textgold.gold, attributes, deprel), raw = True)
````

````python
def joint_table(layers, attributes, deprel):
    table_elements = [css(), event_handler_code(), "<table><tr>"]
    for attribute in attributes:
        for i, layer in enumerate(layers):
            if attribute == "head" or attribute == "deprel":
                table_elements.extend(table_column(layer, attribute, deprel, i + 1))
            elif i == 0:
                table_elements.extend(table_column(layer, attribute, deprel))
    table_elements.append("</tr></table><button type=\"button\">save</button>")
    return "".join(table_elements)
````

````python
display_html(joint_table([textgold.gold, textmalt.malt], attributes, deprel), raw = True)
````
